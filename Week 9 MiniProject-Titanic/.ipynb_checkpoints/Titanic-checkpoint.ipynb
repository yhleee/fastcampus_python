```python
import pandas as pd

titanic = pd.read_csv("dataset/titanic_dataset.csv")
answer = pd.read_csv("dataset/titanic_answer.csv")
```

```python
titanic.head(10)
```

```python
# cabin, age, body, home.dest : 결측데이터 처리
titanic = titanic.drop([ 'body', 'home.dest'], axis=1)
#titanic['age'] = titanic['age'].fillna(titanic['age'].mean())

# name, ticket : 불필요한 feature 제거
titanic = titanic.drop(['name', 'ticket'], axis=1)

# sex : male은 0, female은 1로 변환
titanic.loc[titanic['sex']=='male', 'sex'] = 0
titanic.loc[titanic['sex']=='female', 'sex'] = 1

# Age 결측값 보정 
titanic['child'] =0
titanic.loc[titanic['age']<18,'child']=1
titanic = titanic.drop(['age'], axis=1)

#cabin
titanic['cabin'] = titanic['cabin'].fillna('X')
titanic['cabin'] = titanic['cabin'].apply(lambda x:x[0])
df_one_hot_encoded1 = pd.get_dummies(titanic.cabin)
titanic = pd.concat([titanic, df_one_hot_encoded1], axis=1)
titanic = titanic.drop(['cabin'], axis=1)

#name 
```

```python
#fare 
def fareclassifier(fare):
    if fare <= 8.0:
        return 0
    elif 8 < fare <= 14:
        return 1
    elif 14 < fare <= 30:
        return 2
    else:
        return 3

titanic['fare'] = titanic['fare'].apply(fareclassifier)
```

```python


# embark : 2개의 결측값은 최빈값으로 대체하고, one hot encoding 적용
# 3개
embarked_mode = titanic['embarked'].value_counts().index[0]
titanic['embarked'] = titanic['embarked'].fillna(embarked_mode)

df_one_hot_encoded = pd.get_dummies(titanic.embarked)
titanic = pd.concat([titanic, df_one_hot_encoded], axis=1)
titanic = titanic.drop(['embarked'], axis=1)

# train data, test data 분리
from sklearn.model_selection import train_test_split
train, test = train_test_split(titanic, test_size=0.3, random_state=0)

train_y = train['survived']
train_x = train.drop(['survived'], axis=1)

test_y = test['survived']
test_x = test.drop(['survived'], axis=1)
```

```python
train_x.head(10)
```

####  LogistricRegression

```python
from sklearn.linear_model import LogisticRegression
lrm = LogisticRegression()
lrm.fit(train_x, train_y)
pred_y = lrm.predict(test_x)

from sklearn.metrics import precision_score, recall_score, f1_score
from sklearn.metrics import accuracy_score

print("accuracy: %.2f" %accuracy_score(test_y, pred_y))
print("Precision : %.3f" % precision_score(test_y, pred_y))
print("Recall : %.3f" % recall_score(test_y, pred_y))
print("F1 : %.3f" % f1_score(test_y, pred_y))
```

#### RandomForest

```python

from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(max_depth=3, random_state=0)
clf.fit(train_x, train_y)
Y_pred = clf.predict(test_x)
clf.score(test_x, test_y)

from sklearn.metrics import precision_score, recall_score, f1_score
from sklearn.metrics import accuracy_score

print("accuracy: %.2f" %accuracy_score(test_y, Y_pred))
print("Precision : %.3f" % precision_score(test_y, Y_pred))
print("Recall : %.3f" % recall_score(test_y, Y_pred))
print("F1 : %.3f" % f1_score(test_y, Y_pred))
```

#### SVC 

```python
from sklearn.svm import SVC
clf = SVC(kernel='linear', C=1.0, random_state=0)
clf.fit(train_x, train_y)
pred_y = clf.predict(test_x)

from sklearn.metrics import precision_score, recall_score, f1_score
from sklearn.metrics import accuracy_score

print("accuracy: %.2f" %accuracy_score(test_y, pred_y))
print("Precision : %.3f" % precision_score(test_y, pred_y))
print("Recall : %.3f" % recall_score(test_y, pred_y))
print("F1 : %.3f" % f1_score(test_y, pred_y))
```

#### 의사결정나무

```python
from sklearn.tree import DecisionTreeClassifier

clf2 = DecisionTreeClassifier() # 의사결정나무
clf2 = clf2.fit(train_x, train_y)

```

```python
pred_y = clf2.predict(test_x)

print("accuracy: %.2f" %accuracy_score(test_y, pred_y))
print("Precision : %.3f" % precision_score(test_y, pred_y))
print("Recall : %.3f" % recall_score(test_y, pred_y))
print("F1 : %.3f" % f1_score(test_y, pred_y))
```
